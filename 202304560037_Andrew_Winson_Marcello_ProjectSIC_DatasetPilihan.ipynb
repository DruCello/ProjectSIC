**2.1 Penanganan Nilai Hilang:** Pilih, justifikasikan, dan implementasikan strategi penanganan nilai hilang. Verifikasi hasilnya.

```
import pandas as pd

df = pd.read_csv("/content/dataset_notes_parfum_siang_malam.csv")

# Cek nilai hilang
missing = df.isnull().sum()

# Imputasi numerik
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Imputasi kategorikal
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# Verifikasi
df.isnull().sum()
```

**1. Deteksi Missing Values**

Langkah pertama adalah mendeteksi apakah ada nilai kosong (`NaN`) dalam dataset menggunakan `.isnull().sum()`.

Hasil analisis menunjukkan bahwa terdapat nilai kosong di beberapa kolom numerik dan kategorikal, yang perlu ditangani agar tidak mengganggu proses pemodelan.

**2. Strategi Penanganan**

Untuk kolom numerik: Digunakan imputasi dengan mean (rata-rata).

Untuk kolom kategorikal: Digunakan imputasi dengan modus (nilai yang paling sering muncul).

**3. Mengapa Mean dan Modus?**

**Mean** cocok digunakan pada data numerik yang relatif normal tanpa outlier ekstrem. Ini membuat distribusi tetap representatif dan tidak terlalu condong.

**Modus** ideal untuk data kategorikal karena menggantikan nilai kosong dengan kategori yang paling umum muncul, sehingga tidak mengganggu pola distribusi.

**4. Implementasi**

**Imputasi numerik**

`df[num_cols] = df[num_cols].fillna(df[num_cols].mean())`

**Imputasi kategorikal**

`df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])`

**5. Verifikasi**

Setelah proses imputasi, dicek kembali apakah masih ada nilai kosong.

Hasilnya:

Semua nilai `NaN` berhasil diatasi, dan tidak ada nilai kosong tersisa.

**Kesimpulan:**

Dataset telah dibersihkan dari missing values.

Data numerik dan kategorikal kini lengkap, dan siap digunakan untuk proses analisis atau pelatihan model.

Proses ini penting untuk menjaga keakuratan hasil dan menghindari error teknis dalam proses machine learning.

df_scaled berisi data dengan semua fitur numerik asli yang telah diskalakan menggunakan metode Min-Max Scaling ke rentang [0,1].

**Justifikasi Scaling**

**Fitur yang perlu diskalakan:** Semua fitur numerik kontinu seperti Skor_Siang dan Skor_Malam, agar berada dalam rentang [0,1] dan dapat dibandingkan secara seragam.

Kenapa MinMaxScaler?

Ideal untuk algoritma sensitif terhadap skala, seperti KNN, SVM, dan model berbasis Gradient Descent yang bergantung pada nilai absolut.

Menjaga bentuk distribusi asli tanpa mengubah distribusi menjadi standar seperti StandardScaler.

Memudahkan interpretasi, karena nilai tetap dalam rentang yang konsisten dan tidak menyebabkan perubahan drastis dalam data.

Sesuai untuk data dengan batasan nilai tetap, seperti skor preferensi parfum (0-5), sehingga tetap mempertahankan proporsi antar nilai.

Catatan penting (best practice)

Scaling seharusnya dilakukan setelah split data menjadi train dan test untuk menghindari data leakage.

fit_transform() hanya diterapkan ke data latih, sedangkan transform() diterapkan ke data uji agar model tidak "mengintip" informasi dari data uji.
